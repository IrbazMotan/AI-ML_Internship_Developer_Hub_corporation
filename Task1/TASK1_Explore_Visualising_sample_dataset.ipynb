```python
#importing library
import matplotlib.pyplot as plt
```

```python
#ploting scatter graph so visualise iris sepal length and width 
plt.figure(figsize=(8,4))
plt.scatter(df["SepalLengthCm"], df["SepalWidthCm"], color="yellow")
plt.title("Iris:Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)
plt.show()
```

```python
#ploting histogram legths with its frequency  
plt.figure(figsize=(6,4))
plt.hist(df["SepalLengthCm"], bins=10, color="green", edgecolor="darkblue")
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# mimporting sea born library 
import seaborn as sns
# plotting box plot for all column rather than id column
plt.figure(figsize=(10, 6))

df.drop(columns=['Id']).boxplot()

plt.title("Boxplot of All Numerical Features")
plt.ylabel("Value (cm)")
plt.grid(True)

plt.show()
```
